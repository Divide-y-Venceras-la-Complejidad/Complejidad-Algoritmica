```python
def find(text, word):
  n = len(text)
  l = len(word)
  resp = []
  for i in range(n - l):
    if text[i:i+l] == word:
      resp.append(i)

  return resp

find("abracadabracalamazoo", "rac")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
board = [3, 6, 1, 7, 2, 5, 0, 4]
for row, col in enumerate(board):
  print(f"Hay una reina en la fila {row} y la columna {col}")
```

```python
def draw(board):
  n = len(board)
  b = np.zeros((n, n, 3), dtype=int)
  b += [255, 100, 20]
  b[::2, ::2] = [255, 220, 150]
  b[1::2, 1::2] = [255, 220, 150]

  _, ax = plt.subplots()
  ax.imshow(b)

  for row, col in enumerate(board):
    ax.text(col, row, u"\u265b", fontsize=200/n, va="center", ha="center")
  
  ax.set(xticks=[],yticks=[])
  ax.axis("image")
```

```python
draw(board)
```

```python
def valid(board, row, col):
  n = len(board)
  for row_i in range(row):
    col_i = board[row_i]
    delta = row - row_i
    if col in [col_i, col_i - delta, col_i + delta]:
      return False
  
  return True
```

```python
board = [0, 0, -1, -1]
assert valid(board, 1, 0 ) == False
draw(board)
```

```python
board = [1, 3, 0, -1]
assert valid(board, 2, 0 ) == True
draw(board)
```

```python
board = [1, 3, 2, -1]
assert valid(board, 2, 2) == False
draw(board)
```

```python
def nqueens(board, row):
  n = len (board)
  if row == n:
    txtdraw(board)
  else:
    for col_i in range(n):
      if valid(board, row, col_i):
        board[row] = col_i
        nqueens(board, row + 1)
```

```python
def txtdraw(board):
  n = len(board)
  print("+---"*n + "+")
  for i in range(n):
    for j in range(n):
      print("| " + ("Q" if board[i] == j else " ") + " ", end="")
    print("|")
    print("+---"*n + "+")
```
